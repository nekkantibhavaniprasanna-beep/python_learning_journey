```python
import pandas as pd

# -------------------------------
# 1. String Operations
# -------------------------------

names = pd.Series([
    " Alice ",
    "Bob",
    "Charlie",
    "alice",
    "Bob"
])

print("Original Names:")
print(names)

print("\nLowercase:")
print(names.str.lower())

print("\nUppercase:")
print(names.str.upper())

print("\nRemove Spaces:")
print(names.str.strip())

print("\nContains 'a':")
print(names.str.lower().str.contains("a"))

print("\nReplace 'Alice' with 'Alicia':")
print(names.str.strip().str.replace("Alice", "Alicia"))


# -------------------------------
# 2. Date Operations
# -------------------------------

dates = pd.Series([
    "2024-01-15",
    "2025-06-20",
    "2026-12-31"
])

dates = pd.to_datetime(dates)

print("\nDates:")
print(dates)

print("\nYear:")
print(dates.dt.year)

print("\nMonth:")
print(dates.dt.month)

print("\nDay:")
print(dates.dt.day)


# -------------------------------
# 3. Boolean Filtering
# -------------------------------

marks = pd.Series([45, 72, 88, 30, 95, 60])

print("\nOriginal Marks:")
print(marks)

print("\nMarks greater than 50:")
print(marks[marks > 50])

print("\nMarks less than 50:")
print(marks[marks < 50])


# -------------------------------
# 4. Duplicate Handling
# -------------------------------

students = pd.Series([
    "Rahul",
    "Priya",
    "Rahul",
    "Kiran",
    "Priya",
    "Anjali"
])

print("\nStudent Names:")
print(students)

print("\nDuplicate Values:")
print(students.duplicated())

print("\nAfter Removing Duplicates:")
print(students.drop_duplicates())
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Name": ["Rahul", "Priya", "Kiran", "Anjali", "Rahul"],
    "Age": [20, np.nan, 19, 22, 20],
    "Marks": [85, 90, np.nan, 88, 85],
    "City": ["Hyderabad", "Vizag", "Vizag", None, "Hyderabad"]
})

print(df)
print(df.isnull())
print(df.isnull().sum())
print(df.fillna(0))
df["Age"] = df["Age"].fillna(df["Age"].mean())
df["Marks"] = df["Marks"].fillna(df["Marks"].median())
df["City"] = df["City"].fillna(df["City"].mode()[0])
print(df.dropna())
print(df.dropna(axis=1))
print(df.duplicated())
print(df.duplicated().sum())
print( df.drop_duplicates())
df["Age"] = df["Age"].astype(int)
print(df.dtypes)
print(df.rename(columns={
    "Marks": "Score",
    "City": "Location"
}))
df["City"] = df["City"].replace("Vizag", "Visakhapatnam")
df["City"] = df["City"].replace({
    "Hyderabad": "HYD",
    "Visakhapatnam": "VSKP"
})
print(df["City"])
```

```python
import pandas as pd

df = pd.DataFrame({
    "Name":["Rahul","Priya","Kiran","Anjali","Ravi","Sneha"],
    "Age":[20,22,19,21,20,22],
    "Department":["AI","CSE","AI","ECE","CSE","AI"],
    "Marks":[85,90,78,88,70,95],
    "City":["Hyderabad","Vizag","Hyderabad","Vijayawada","Vizag","Hyderabad"]
})

print(df)

print(df.query("Marks>80"))
print(df.query("Age>=20"))
print(df.query("Department == 'AI' "))
print(df.sort_values("Marks",ascending=False)) # Descending
print(df["Department"].value_counts())
print(df.groupby("Department")["Marks"].mean())
print(df.groupby("Department")["Marks"].max())
print(df.groupby("Department")["Marks"].agg(["max","min","mean","count"]))
print(df["Marks"].apply(lambda x: x+5))
df["Grade"] = df["Marks"].apply(lambda x: "A" if x>=90 else ("B" if x>=80 else ("C" if x>=70 else "D")))
print(df["Grade"])
df["Deptcode"]=df["Department"].map({
    "AI" : 1,
    "CSE": 2,
    "ECE" : 3
})
print(df["Deptcode"])
print(df["City"].unique())
print(df["City"].value_counts())
```

```python
import pandas as pd

students = pd.DataFrame({
    "ID":[101,102,103,104],
    "Name":[" Rahul ","Priya "," Kiran","Anjali"],
    "Department":["AI","CSE","ECE","AI"],
    "Joining_Date":["2023-06-15","2022-08-20","2023-01-10","2021-12-05"]
})

marks = pd.DataFrame({
    "ID":[101,102,103,104],
    "Marks":[85,92,78,88]
})
```

```python
print(students["Name"].str.strip())
print(students["Name"].str.upper())
students["dept_code"]=students["Department"].map({
    "AI":1,
    "CSE":2,
    "ECE":3
})
print(students["dept_code"])
pd.get_dummies(
    students,
    columns=["Department"]
)
print(students.merge(marks,on="ID"))
print(students)
students["Joining_Date"]= pd.to_datetime(
    students["Joining_Date"]
)
print(students["Joining_Date"].dt.year)
print(students["Joining_Date"].dt.month)
print(students["Joining_Date"].dt.day_name)
print(students["Name"].str.startswith("R"))


```
