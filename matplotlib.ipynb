```python
# Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Create Sample Dataset
data = {
    "Name": ["Ram", "Sita", "Ravi", "Priya", "Kiran"],
    "Age": [18, 19, 20, 21, 22],
    "Marks": [85, 92, 75, 88, 95],
    "Gender": ["M", "F", "M", "F", "M"]
}

df = pd.DataFrame(data)

# Step 3: Display Dataset
print(df)

# Step 4: Basic Information
print("\nDataset Info:")
print(df.info())

print("\nSummary:")
print(df.describe())

# Step 5: Histogram
plt.figure(figsize=(5,4))
plt.hist(df["Marks"])
plt.title("Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()

# Step 6: Scatter Plot
plt.figure(figsize=(5,4))
plt.scatter(df["Age"], df["Marks"])
plt.title("Age vs Marks")
plt.xlabel("Age")
plt.ylabel("Marks")
plt.show()

# Step 7: Bar Chart
plt.figure(figsize=(5,4))
plt.bar(df["Name"], df["Marks"])
plt.title("Marks of Students")
plt.xlabel("Name")
plt.ylabel("Marks")
plt.show()

# Step 8: Pie Chart
plt.figure(figsize=(5,4))
df["Gender"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)
plt.title("Gender Distribution")
plt.ylabel("")
plt.show()

# Step 9: Box Plot
plt.figure(figsize=(5,4))
plt.boxplot(df["Marks"])
plt.title("Marks Box Plot")
plt.show()

# Step 10: Correlation
print("\nCorrelation Matrix:")
print(df.corr(numeric_only=True))
```

```python
import matplotlib.pyplot as plt

# ----------------------------
# 1. Training Loss
# ----------------------------
epochs = [1, 2, 3, 4, 5]

loss = [0.9, 0.7, 0.5, 0.3, 0.2]

plt.figure(figsize=(6,4))
plt.plot(epochs, loss, marker='o', color='red')
plt.title("Training Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.grid()
plt.show()


# ----------------------------
# 2. Training vs Validation Accuracy
# ----------------------------
train_accuracy = [70, 78, 85, 90, 95]
validation_accuracy = [68, 75, 82, 88, 92]

plt.figure(figsize=(6,4))
plt.plot(epochs, train_accuracy, marker='o', label="Train Accuracy")
plt.plot(epochs, validation_accuracy, marker='s', label="Validation Accuracy")
plt.title("Accuracy Comparison")
plt.xlabel("Epoch")
plt.ylabel("Accuracy (%)")
plt.legend()
plt.grid()
plt.show()


# ----------------------------
# 3. Actual vs Predicted
# ----------------------------
actual = [100, 120, 140, 160, 180]
predicted = [98, 118, 145, 158, 182]

plt.figure(figsize=(6,4))
plt.scatter(actual, predicted, color="green")
plt.title("Actual vs Predicted")
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.grid()
plt.show()


# ----------------------------
# 4. Feature Importance
# ----------------------------
features = ["Age", "Salary", "Experience", "Education"]
importance = [0.20, 0.35, 0.30, 0.15]

plt.figure(figsize=(6,4))
plt.barh(features, importance, color="orange")
plt.title("Feature Importance")
plt.xlabel("Importance")
plt.show()
```
