```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('titanic')
print(df.head())
print(df.tail())
```

```python
sns.set_style("darkgrid")
sns.set_palette("rainbow")
sns.countplot(x="sex", data=df)
sns.set_theme()

plt.show()
```

```python
sns.countplot(x="class",data=df)
plt.show()
```

```python
sns.barplot(x="sex", y="age", data=df)
plt.show()
```

```python
sns.barplot(x="class", y="fare", data=df)
```

```python
sns.histplot(df["age"],bins=20)
plt.show()
```

```python
sns.kdeplot(df["age"],fill=True)
plt.show()
```

```python
sns.histplot(df["age"], kde=True)
plt.show()
```

```python
plt.figure(figsize=(3,3))

sns.countplot(x="sex", data=df)

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(x="sex", data=df)

plt.title("Male vs Female")
plt.xlabel("Gender")
plt.ylabel("Count")

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("titanic")

sns.scatterplot(x="age", y="fare", hue="sex",style="class",data=df)
plt.show()
```

```python
sns.scatterplot(
    x="age",
    y="fare",
    size="fare",
    data=df
)
```

```python
sns.scatterplot(
    x="age",
    y="fare",
    hue="sex",
    style="class",
    size="fare",
    data=df
)

plt.show()
```

```python
sns.lineplot(
    x="age",
    y="fare",
    data=df
)
```

```python
tips = sns.load_dataset("tips")

sns.lineplot(
    x="size",
    y="tip",
    data=tips
)

plt.show()
```

```python
sns.regplot(
    x="age",
    y="fare",
    data=df
)

plt.show()
```

| Function        | Purpose                                      |
| --------------- | -------------------------------------------- |
| `scatterplot()` | Relationship between two numerical variables |
| `lineplot()`    | Trends over ordered values                   |
| `regplot()`     | Scatter plot with regression line            |
| `lmplot()`      | Regression plot with grouping                |
| `hue`           | Different colors for categories              |
| `style`         | Different marker shapes                      |
| `size`          | Marker size based on a numerical variable    |


```python
sns.lmplot(
    x="age",
    y="fare",
    hue="sex",
    data=df
)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("titanic")

sns.boxplot(y="age", data=df)

plt.show()
```

```python
sns.violinplot(y="age", data=df)

plt.show()
```

```python
sns.stripplot(
    y="age",
    data=df
)

plt.show() #Shows every data point.
```

```python
sns.swarmplot(
    y="age",
    data=df
) #Dots never overlap.
```

```python
sns.boxplot(
    x="class",
    y="age",
    data=df
)
```

```python
sns.violinplot(
    x="class",
    y="age",
    data=df
)
```

```python
sns.stripplot(
    x="class",
    y="age",
    data=df
)
```
