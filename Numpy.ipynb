```python
import numpy as np

# Original Array
arr = np.array([12, 25, 37, 48, 59, 60, 71, 82, 93, 100])

print("Original Array:")
print(arr)

# 1. Mean
print("\n1. Mean:")
print(np.mean(arr))

# 2. Median
print("\n2. Median:")
print(np.median(arr))

# 3. Standard Deviation
print("\n3. Standard Deviation:")
print(np.std(arr))

# 4. Variance
print("\n4. Variance:")
print(np.var(arr))

# 5. Even Numbers
print("\n5. Even Numbers:")
print(arr[arr % 2 == 0])

# 6. Odd Numbers
print("\n6. Odd Numbers:")
print(arr[arr % 2 != 0])

# 7. Numbers Greater Than Mean
mean = np.mean(arr)
print("\n7. Numbers Greater Than Mean:")
print(arr[arr > mean])

# 8. Replace Odd Numbers with -1
new_arr = arr.copy()
new_arr[new_arr % 2 != 0] = -1

print("\n8. Replace Odd Numbers with -1:")
print(new_arr)

# 9. Index of Maximum Value
print("\n9. Index of Maximum Value:")
print(np.argmax(arr))

# 10. Min-Max Normalization
normalized = (arr - arr.min()) / (arr.max() - arr.min())

print("\n10. Min-Max Normalization:")
print(normalized)
```

```python
import numpy as np

# Dataset
data = np.array([
    [23, 45000],
    [25, 50000],
    [30, 60000],
    [35, 70000],
    [40, 80000],
    [28, 55000],
    [32, 65000],
    [26, 52000],
    [45, 90000],
    [38, 75000]
])

print("Original Dataset:")
print(data)

# ---------------------------------------
# 31. Replace every age below 30 with 30
# ---------------------------------------
age_data = data.copy()
age_data[age_data[:, 0] < 30, 0] = 30

print("\n31. Replace Age < 30 with 30:")
print(age_data)

# ---------------------------------------
# 32. Clip salaries between 50000 and 80000
# ---------------------------------------
clip_data = data.copy()
clip_data[:, 1] = np.clip(clip_data[:, 1], 50000, 80000)

print("\n32. Clip Salaries (50000 - 80000):")
print(clip_data)

# ---------------------------------------
# 33. Check if any employee earns > 85000
# ---------------------------------------
print("\n33. Any Salary > 85000?")
print(np.any(data[:, 1] > 85000))

# ---------------------------------------
# 34. Check if all employees earn > 40000
# ---------------------------------------
print("\n34. All Salaries > 40000?")
print(np.all(data[:, 1] > 40000))

# ---------------------------------------
# 35. Find Unique Ages
# ---------------------------------------
print("\n35. Unique Ages:")
print(np.unique(data[:, 0]))

# ---------------------------------------
# 36. Count Employees Older Than 30
# ---------------------------------------
print("\n36. Employees Older Than 30:")
print(np.count_nonzero(data[:, 0] > 30))

# ---------------------------------------
# 37. Reshape Salary into (10,1)
# ---------------------------------------
salary = data[:, 1].reshape(10, 1)

print("\n37. Salary Reshaped (10x1):")
print(salary)

# ---------------------------------------
# 38. Flatten Salary Array
# ---------------------------------------
print("\n38. Flatten Salary:")
print(salary.flatten())

# ---------------------------------------
# 39. Dot Product of Salary and Weight Vector
# ---------------------------------------
weights = np.ones(10)

dot_product = np.dot(data[:, 1], weights)

print("\n39. Dot Product:")
print(dot_product)

# ---------------------------------------
# 40. Mean Salary Using Two Methods
# ---------------------------------------
mean1 = np.mean(data[:, 1])
mean2 = data[:, 1].mean()

print("\n40. Mean Using np.mean():")
print(mean1)

print("\nMean Using arr.mean():")
print(mean2)

print("\nAre Both Equal?")
print(mean1 == mean2)
```
